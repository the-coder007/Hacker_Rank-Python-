```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if year%4==0:
        if year%400:
            leap=True
        elif year%100==0:
            if year==2000 or year==2400:
                leap=True
            else:
                leap=False
        elif year%4==0:
            leap=True
    return leap
```

```python
is_leap(2000)
```
